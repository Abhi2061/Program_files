```python
import matplotlib.pyplot as plt
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_title('Directions')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.plot([5],[5],'o')

ax.text(4.5, 0.5, 'North', verticalalignment = 'bottom', color = 'red', fontsize = 15, bbox = {})
ax.text(4.5, 9, 'South', verticalalignment = 'bottom', color = 'red', fontsize = 15, bbox = {})
ax.text(8, 4.8, 'East', verticalalignment = 'bottom', color = 'red', fontsize = 15, bbox = {})
ax.text(1.2, 4.8, 'West', verticalalignment = 'bottom', color = 'red', fontsize = 15, bbox = {})

ax.annotate('', xy = (5,1), xytext=(5,5), arrowprops = dict(facecolor = 'yellow', shrink = 0.05))
ax.annotate('', xy = (5,9), xytext=(5,5), arrowprops = dict(facecolor = 'blue', shrink = 0.05))
ax.annotate('', xy = (2,5), xytext=(5,5), arrowprops = dict(facecolor = 'green', shrink = 0.05))
ax.annotate('', xy = (8,5), xytext=(5,5), arrowprops = dict(facecolor = 'red', shrink = 0.05))

ax.axis([1,10,0,10])
plt.show()
```
