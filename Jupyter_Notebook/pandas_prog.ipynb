```python
data = {'Name' : ['Tom', 'Nick', 'Krish', 'Jack'],
    'Age' : [20, 21, 19, 18],
    'Qualification' : ['MCA', 'MSC', 'BCA', 'BBA']}

df = pd.DataFrame(data)

print('Original values :')
display(df)

print('Selected Columns :')
display(df[['Name', 'Qualification']])
```

```python
print('Before Changing : ')
display(df)

df.loc[df.Name == 'Krish', 'Age'] = 25

print('After Changing : ')
display(df)
```

```python
print('Original values :')
display(df)

print('Selected Rows : ')
display(df.loc[(df.Age > 19)])
```

```python
print("Original Data")
display(df)

# sum of items
print("sum of items , axis = 0")
display(df.sum(axis=0, skipna=True))

print("sum of items , axis = 1")
display(df.sum(axis=1, skipna=True))

# minimum and maximum of items
print("minimum and maximum of items")
display(df.min(axis=0, skipna=True))
display(df.max(axis=0, skipna=True))

# mean of items
print("mean of items")
display(df.mean(axis=0, skipna=True))

# aggregate of all
print("aggregate of all")
display(df.agg(['sum', 'min', 'max']))
```
